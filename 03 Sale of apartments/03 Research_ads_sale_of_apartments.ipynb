```python
def info_row(df):
    '''
        Input: df - таблица pd.DataFrame.
        Output: None.
        Description: Функция первичного знакомства с данными.

    '''
    
    display(df.head(10))
    display(df.sample(10))
    print('Общая информация:')
    display(df.info())
    print('Общая статистика:')
    display(df.describe())
    display(df.describe(include = ['object']))
    print('Дубликаты:')
    display(df.duplicated().sum())
    print(f'{df.duplicated().sum()/len(df):.1%}')
    print('Пропуски:')
    for col in df.columns:
        pct_missing = np.mean(df[col].isna())
        print('{} - {}%'.format(col, round(pct_missing*100,2)))
    print('Пропуски на Heatmap:')    
    columns = df.columns[:]
    colours = ['#ADD8E6', '#FFFACD'] 
    plt.figure(figsize=(20, 20))
    sns.heatmap(data[columns].isnull(), cmap=sns.color_palette(colours))
    plt.show()
    print('Уникальные значения столбцов:')
    for row in df.columns:
        print('Уникальные значения столбца', row)
        print(df[row].unique())
```

Можно попробовать соотнести пропуски с населенным пунктом или отдаленности от центра, для того, чтобы заполнить аналогичным значением. <font color='red'>ВАЖНЫЕ ДАННЫЕ.</font>*

### Задача 3  
*Уберите редкие и выбивающиеся значения. Опишите, какие особенности обнаружили.*

```python
mean_data = data.copy(deep=True)
mean_data.head()

# Создадим mean_data, в который будем вносить отфильтрованные значения.
```

*Откинуть выбросы необходимо в столбцах:
`last_price`,`total_area`,`kitchen_area`,`living_area`,`days_exposition`,`ceiling_height`,`rooms`, так как данные столбцы важны, для дальнейшего анализа данных.*

```python

def quantile_stat(df,row):
    '''
        Input: df,row - таблица pd.DataFrame и наименование колонки из
               списка columns_emissions.
        Output: df - таблица pd.DataFrame, очищенная от выбросов.
        Description: Функция вычисляет квантили и очищает данные от выбросов.

    '''
    q75 = df[row].quantile(3/4)
    q25 = df[row].quantile(1/4)
    qmax = df[row].max()
    qmin = df[row].min()
    intr_q = q75-q25
    max_q = q75+(1.5*intr_q)
    min_q = q25-(1.5*intr_q)
    df_q = df.loc[(df[column] <= max_q) & (df[column] >= min_q), column]
    if max_q > qmax:
        max_q = qmax 
    if min_q < qmin:
        min_q = qmin
    return df_q 


columns_emissions = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 
                     'ceiling_height', 'rooms']
for column in columns_emissions:
    mean_data[column] = quantile_stat(mean_data, column) 
```
