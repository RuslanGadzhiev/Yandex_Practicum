```python
def drop_dup(df):
    '''
        Input: df - таблица pd.DataFrame.
        Output: None.
        Description: Функция удаляет дубликаты, оставляя первое вхождение.

    '''
    df.drop_duplicates(subset='id', keep='first', inplace=True, ignore_index=False)
    df.reset_index(drop=True)
    
```

### Деление данных на обучающую и валидационную выборки

```python
def split_df(features,target,color,region):
    '''
        Input: features,target - признаки и целевой признак, color - цвет вывода,
               region - наименование региона.
        Output: features,target - признаки и целевой признак.
        Description: Функция делит данные на обучающую и валидационную выборки.

    '''    
    features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, train_size=0.75,random_state=12345)
    
    display(HTML(f"<font color={color}>Целевые признаки обучающей и валидационной сборок {region}: \
    {target_train.shape},{target_valid.shape}</font>"))
    display(HTML(f"<font color={color}>Признаки обучающей и валидационной сборок {region}: \
    {features_train.shape},{features_valid.shape}</font>"))

    return features_train,features_valid,target_train,target_valid
```

```python
features_train_f, features_valid_f, target_train_f, target_valid_f = split_df(features_f,
                                                                              target_f,'orange','(Первый регион)')
```

```python
features_train_s, features_valid_s, target_train_s, target_valid_s = split_df(features_s,
                                                                              target_s,'blue','(Второй регион)')
```

```python
features_train_t, features_valid_t, target_train_t, target_valid_t = split_df(features_t,
                                                                              target_t,'green','(Третий регион)')
```

### Обучение модели и получение предсказаний на валидационной выборке

```python
def model_linear(features,target,features_train, features_valid, target_train, target_valid, color, region):
    '''
        Input: features,target,features_train, features_valid, target_train, target_valid - 
               признаки и целевой признак, признаки и целевые признаки обучающей и валидационной выборок;
               color - цвет вывода, region - наименование региона.
        Output: predictions_valid,result_rmse,model - предсказания, значение RMSE и обученная модель.
        Description: Функция обучает модель LinearRegression 
                     и вычисляет средний запас предсказанного сырья и RMSE.

    ''' 
    model = LinearRegression()
    model.fit(features_train,target_train)
    predictions_valid = model.predict(features_valid)
    result_rmse = mean_squared_error(target_valid,predictions_valid)**0.5
    display(HTML(f"<font color = {color}>Средний запас предсказанного сырья {region}:\
    {predictions_valid.mean():.2f} тысяч баррелей</font>"))
    display(HTML(f"<font color = {color}>Коэффициент RMSE модели {region}:\
    {result_rmse:.2f}</font>"))
    
    return predictions_valid,result_rmse,model
```

```python
predictions_valid_f,result_rmse_f,model_f = model_linear(features_f,target_f,features_train_f, 
                                                features_valid_f, target_train_f, 
                                                         target_valid_f,'orange','(Первый регион)')
```

```python
BUDGET_ON_REGION = 10000000000
WELLS_REGION = 500
BEST_WELLS = 200
BUDGET_ON_BEST_UNIT = BUDGET_ON_REGION/BEST_WELLS
REVENUE_PER_BAR = 450000
RISK_FACTOR = 0.025

# Вводим данные для расчетов.
# BUDGET_ON_REGION - бюджет на разработку скважин в регионе — 10 млрд рублей.
# WELLS_REGION - количество скважин, иследуемых при разведке региона.
# BEST_WELLS - количество отобранных скважин с наибольшими объемами сырья.
# BUDGET_ON_BEST_UNIT - бюджет на разработку одной из отобранных скважин.
# REVENUE_PER_BAR - доход с каждой единицы продукта.
# RISK_FACTOR - допустимая вероятность убытков.
```

### Вычисление объема сырья для безубыточной разработки новой скважины

*ROI (от англ. return on investments, «возврат на инвестиции»).   
ROI скважины = (Доход от скважины - Инвестиции)/Инвестиции. Если ROI равен 0, то вложения окупились, но прибыли нет, а значит это и есть точка безубыточности.*  
* *В таком случае Доход от скважины = Инвестиции.*  
* *Доход от скважины = Объем сырья х Прибыль от продажи единицы сырья(баррель, тыс.).*  
* *Значит объем сырья = Инвестиции/Прибыль от продажи единицы сырья.*  
* *Инвестиции на разработку одной из 200 лучших скважин - `BUDGET_ON_BEST_UNIT`.*  
* *Прибыль от продажи единицы сырья - `REVENUE_PER_BAR`.*

```python
VOLUME_OF_RAW = BUDGET_ON_BEST_UNIT/REVENUE_PER_BAR
display(HTML(f"<font color = 'violet'>Достаточный объём сырья для безубыточной разработки новой скважины: \
    {VOLUME_OF_RAW:.2f} тысяч баррелей</font>"))
```

```python
def promising_wells(df,color,region):
    '''
        Input: df - таблица pd.DataFrame, color - цвет вывода, region - наименование региона.
        Output: promising_percent - процент перспективных скважин.
        Description: Функция подсчета процента безубыточных скважин в регионе.

    '''
    count = 0 
    for i in df['product']:
        if i > VOLUME_OF_RAW:
            count += 1 
    promising_percent = count/len(df['product'])
    display(HTML(f"<font color = {color}>Процент перспективных скважин в {region}:\
    {promising_percent:.2%}</font>"))
    
    return promising_percent
        
```

## Расчёт прибыли и рисков 

### Функция расчета прибыли по предсказаниям модели

```python
probabilities_valid_f = pd.Series(predictions_valid_f,index=target_valid_f.index)
probabilities_valid_s = pd.Series(predictions_valid_s,index=target_valid_s.index)
probabilities_valid_t = pd.Series(predictions_valid_t,index=target_valid_t.index)
```

```python
def profit(target, predictions):
    '''
        Input: target, predictions - целевой признак и предсказания модели.
        Output: quantity_over - общая выручка по двумста лучшим скважинам.
        Description: Функция расчета прибыли по двухста скважинам с самыми большими объемами сырья, 
                     полученными по предсказаниям модели.
                     1) Объединили предсказания и целевой валидационный признак в один df с общими индексами
                     2) Отсортировали двести скважин с самыми большими объемами сырья(по предсказаниям)
                     3) Выбрали соответствующие по индексам скважины в целевом признаке
                     4) Высчитали общую выручку по ним

    '''
    quantity_over = 0
    data_test = pd.concat([predictions, target],axis=1)
    data_test.columns = ['predictions','target']
    
    reg_wells = data_test.sort_values(by = 'predictions',ascending=False).head(BEST_WELLS)
    selected_wells = target[reg_wells.index][:BEST_WELLS]
    for wells in selected_wells:
        quantity_over += (wells - VOLUME_OF_RAW)*REVENUE_PER_BAR
    return quantity_over

```

### Функция расчета прибыли и рисков методом Bootstrap

```python
def profit_and_risks(target,probabilities,color,region):
    '''
        Input: target, probabilities - целевой признак и список вероятностей,
               color - цвет вывода, region - наименование региона.
        Output: mean - средняя выручка, lower,upper - 95% доверительный интервал, risks - вероятность убытков.
        Description: Функция расчета прибыли и рисков методом Bootstrap.
                     1) Создали 1000 рандомных выборок по 500 скважин(исследуемых при разведки региона)
                     2) Выбрали для каждой из выборок 200 лучших скважин, применив функцию revenue
                     3) Нашли среднюю прибыль, 95%-й доверительный интервал и риск убытков

    '''
    state = np.random.RandomState(12345)
    values = []
    
    for i in range(1000):
        
        target_subsample = target.sample(n = WELLS_REGION, replace=True, random_state=state) 
        probs_subsample = probabilities[target_subsample.index]
    
        
        prof = profit(target_subsample, probs_subsample)
        values.append(prof)

    values = pd.Series(values)
    lower = values.quantile(0.025)/10e6
    upper = values.quantile(0.975)/10e6
    mean = values.mean()/10e6
    risks = (values < 0).mean()
    
    display(HTML(f"<font color = {color}>Средняя выручка в {region}:\
    {mean:.2f} млн. руб.</font>"))
    display(HTML(f"<font color = {color}>95% доверительный интервал в {region} от {lower:.2f} до {upper:.2f} \
    млн. руб.</font>"))
    display(HTML(f"<font color = {color}>Вероятность убытков в {region}:\
    {risks:.1%}</font>"))
    
    return mean,lower,upper,risks
```
